```python
import cv2
import numpy as np
import os

# Cargar la imagen principal
def cargar_imagen(nombre_archivo):
    ruta = os.path.join(os.getcwd(), nombre_archivo)
    imagen = cv2.imread(ruta)
    if imagen is None:
        print(f"Error al cargar la imagen: {ruta}")
        exit()
    return imagen

imagen_original = cargar_imagen('C:/Users/Usuario/Downloads/persona.jpg')  # Reemplaza con tu imagen
marca_agua = cargar_imagen('C:/Users/Usuario/Downloads/marca.png')  # Reemplaza con tu marca de agua

# 1. Umbral binario y escala de grises
imagen_gris = cv2.cvtColor(imagen_original, cv2.COLOR_BGR2GRAY)
_, imagen_binaria = cv2.threshold(imagen_gris, 127, 255, cv2.THRESH_BINARY)

# 2. Conversión a HSV y guardado
imagen_hsv = cv2.cvtColor(imagen_original, cv2.COLOR_BGR2HSV)
cv2.imwrite('imagen_hsv.jpg', imagen_hsv)

# 3. Marca de agua (ajustable a diferentes tamaños)
def agregar_marca_agua(imagen, marca_agua, posicion=(10, 10), transparencia=0.5):
    if marca_agua is None:
        print("Error al cargar la imagen de marca de agua.")
        return imagen
    
    factor_escala = min(imagen.shape[1] / marca_agua.shape[1], imagen.shape[0] / marca_agua.shape[0]) / 4
    nuevo_ancho = int(marca_agua.shape[1] * factor_escala)
    nuevo_alto = int(marca_agua.shape[0] * factor_escala)
    marca_agua_redimensionada = cv2.resize(marca_agua, (nuevo_ancho, nuevo_alto))

    roi = imagen[posicion[1]:posicion[1] + nuevo_alto, posicion[0]:posicion[0] + nuevo_ancho]
    roi = cv2.addWeighted(roi, 1 - transparencia, marca_agua_redimensionada, transparencia, 0)
    imagen[posicion[1]:posicion[1] + nuevo_alto, posicion[0]:posicion[0] + nuevo_ancho] = roi
    return imagen

imagen_con_marca_agua = agregar_marca_agua(imagen_original.copy(), marca_agua)

# 4. Combinar imágenes
imagen_pequena = cargar_imagen('C:/Users/Usuario/Downloads/caballo.jpg')  # Reemplaza con tu imagen pequeña
x_offset = imagen_original.shape[1] // 2 - imagen_pequena.shape[1] // 2
y_offset = imagen_original.shape[0] // 2 - imagen_pequena.shape[0] // 2
imagen_combinada = imagen_original.copy()
imagen_combinada[y_offset:y_offset + imagen_pequena.shape[0], x_offset:x_offset + imagen_pequena.shape[1]] = imagen_pequena

# 5. Ruido y operaciones morfológicas
imagen_ruido = imagen_original.copy()
cv2.randn(imagen_ruido, (0, 0, 0), (50, 50, 50))
kernel = np.ones((5, 5), np.uint8)
imagen_cierre = cv2.morphologyEx(imagen_ruido, cv2.MORPH_CLOSE, kernel)
imagen_apertura = cv2.morphologyEx(imagen_ruido, cv2.MORPH_OPEN, kernel)

# 6. Cambiar brillo
imagen_brillo = cv2.convertScaleAbs(imagen_original, alpha=1.5, beta=50)

# 7. Difuminar
imagen_difuminada = cv2.GaussianBlur(imagen_original, (5, 5), 0)

# 8. Filtro Sobel horizontal
imagen_sobel = cv2.Sobel(imagen_gris, cv2.CV_64F, 1, 0, ksize=5)
imagen_sobel = cv2.convertScaleAbs(imagen_sobel)

# 9. Detectar esquinas (Shi-Tomasi)
esquinas = cv2.goodFeaturesToTrack(imagen_gris, 100, 0.01, 10)
esquinas = np.int64(esquinas)
imagen_esquinas = imagen_original.copy()
for esquina in esquinas:
    x, y = esquina.ravel()
    cv2.circle(imagen_esquinas, (x, y), 3, (0, 0, 255), -1)

# Mostrar resultados (opcional)
cv2.imshow('Original', imagen_original)
# ... (mostrar el resto de las imágenes)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import os

# Cargarimágen
imagen_original = cargar_imagen('C:/Users/Usuario/Downloads/persona.jpg')
marca_agua = cargar_imagen('C:/Users/Usuario/Downloads/marca.png')
imagen_pequena = cargar_imagen('C:/Users/Usuario/Downloads/caballo.jpg')

# almacenar
imagen_actual = imagen_original.copy()
imagen_modificada = None
operacion_actual = None

# Función para mostrar la imagen
def mostrar_imagen(imagen, operacion):
    cv2.putText(imagen, operacion, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Imagen', imagen)

while True:
    mostrar_imagen(imagen_actual, operacion_actual if operacion_actual else 'Original')

    tecla = cv2.waitKey(0)

    if tecla == ord('q'):  # Salir al presionar 'q'
        break

    elif tecla == ord('1'):
        # Umbral binario
        imagen_gris = cv2.cvtColor(imagen_original, cv2.COLOR_BGR2GRAY)
        _, imagen_modificada = cv2.threshold(imagen_gris, 127, 255, cv2.THRESH_BINARY)
        operacion_actual = 'Umbral Binario'

    elif tecla == ord('2'):
        # HSV
        imagen_modificada = cv2.cvtColor(imagen_original, cv2.COLOR_BGR2HSV)
        operacion_actual = 'HSV'

    elif tecla == ord('3'):
        # Marca de agua
        imagen_modificada = agregar_marca_agua(imagen_original.copy(), marca_agua)
        operacion_actual = 'Marca de Agua'

    # Actualizar la imagen
    if imagen_modificada is not None:
        imagen_actual = imagen_modificada.copy()
        imagen_modificada = None  # Restablecer para la próxima operación

cv2.destroyAllWindows()
```
